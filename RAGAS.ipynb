```python
import os, openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
```

```python
from datasets import Dataset
from ragas.metrics import faithfulness, answer_relevancy, context_precision, context_recall
from ragas import evaluate

data_samples = {
    'question': ['When was Attention Is All You Need published and by whom?', 'When was Attention Is All You Need  published and by whom?'],
    'answer': ['Attention Is All You Need was published by Taylor Swift in 2017', 'Attention Is All You Need was published by Taylor Swift in 2003'],
    'contexts' : [['Attention Is All You Need was published by Vaswani et al in 2017'], ['Attention Is All You Need was published by Vaswani et al in 2017']],
}
dataset = Dataset.from_dict(data_samples)
```

```python
score = evaluate(dataset, metrics=[faithfulness,answer_relevancy],raise_exceptions=False)
score.to_pandas()
```
